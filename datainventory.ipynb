```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path("data")

if not DATA_DIR.exists():
    raise FileNotFoundError(f"Couldn't find '{DATA_DIR.resolve()}'. Run this from your project root.")

csv_paths = sorted(DATA_DIR.rglob("*.csv"))

print(f"Found {len(csv_paths)} CSV files under {DATA_DIR.resolve()}\n")

rows = []
for p in csv_paths:
    rows.append({
        "relative_path": str(p.relative_to(DATA_DIR)),
        "filename": p.name,
        "size_mb": round(p.stat().st_size / (1024**2), 3),
        "folder": str(p.parent.relative_to(DATA_DIR)) if p.parent != DATA_DIR else ".",
    })

df = pd.DataFrame(rows).sort_values(["folder", "filename"]).reset_index(drop=True)

try:
    from IPython.display import display
    display(df)
except Exception:
    print(df.to_string(index=False))

out_path = Path("data_csv_inventory.csv")
df.to_csv(out_path, index=False)
print(f"\nSaved inventory to: {out_path.resolve()}")

PEEK_COLUMNS = False

if PEEK_COLUMNS:
    schema_rows = []
    for p in csv_paths:
        try:
            sample = pd.read_csv(p, nrows=0)  
            schema_rows.append({
                "relative_path": str(p.relative_to(DATA_DIR)),
                "num_cols": len(sample.columns),
                "columns": ", ".join(sample.columns[:30]) + (" ..." if len(sample.columns) > 30 else ""),
            })
        except Exception as e:
            schema_rows.append({
                "relative_path": str(p.relative_to(DATA_DIR)),
                "num_cols": None,
                "columns": f"ERROR reading header: {e}",
            })
    schema_df = pd.DataFrame(schema_rows)
    try:
        display(schema_df)
    except Exception:
        print(schema_df.to_string(index=False))
```
